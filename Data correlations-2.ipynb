```python
import sqlite3
import pandas as pd

# Establish a connection to the SQLite database
conn = sqlite3.connect('crime_data.db')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Define your SQL query
sql_query = "SELECT * FROM all_burg"

# Execute the query and fetch the results
cursor.execute(sql_query)

# Fetch all the rows from the query result
rows = cursor.fetchall()

# Get the column names from the cursor description
columns = [column[0] for column in cursor.description]

# Create a DataFrame from the fetched rows and column names
df_crimes = pd.DataFrame(rows, columns=columns)

# Close the cursor and connection
cursor.close()
conn.close()

# Display the DataFrame
df_crimes

```

```python
df_crimes.columns
```

```python
df_crimes['year'] =  df_crimes['Month'].str.split('-').str[0]
df_crimes
```

```python
# Assuming you already have a DataFrame named 'df' with a 'LSOA Code' column

# Group the DataFrame by 'LSOA Code' and count the number of occurrences
burglaries_count = df_crimes.groupby(by = ['LSOA code', 'year'] , as_index=False).count()

# Display the resulting DataFrame
print(burglaries_count)
```

```python
burglaries_count
```

```python
burlg_f = burglaries_count[['LSOA code', 'year', 'Latitude']]
```

```python

burlg_f =burlg_f.rename(columns={ 'Latitude': 'count'})
```

```python
burlg_f
```

```python
# Assuming you have a DataFrame named 'df' with columns 'LSOA code', 'year', and 'count'

# Pivot the DataFrame to create columns with the format 'count_{year}'
pivoted_df = burlg_f.pivot(index='LSOA code', columns='year', values='count')

# Rename the new columns with the format 'count_{year}'
pivoted_df.columns = [f'count_{year}' for year in pivoted_df.columns]

# Reset the index to make 'LSOA code' a regular column
pivoted_df = pivoted_df.reset_index()

# Display the transformed DataFrame
pivoted_df
```

```python
df_demographics=df_demographics.rename(columns={ 'Codes': 'LSOA code'})
```

```python
df_demographics
```

```python
demo_complete = df_demographics.merge(pivoted_df, on='LSOA code', how='inner')
demo_complete
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def corr_year(y) :
    year = str(y)
    selected_cols = [col for col in demo_complete.columns if year in col] + ['LSOA code', 'Names']
    df = demo_complete[selected_cols]
    n = 'count_' + year
    # Calculate the correlation between 'count' and each column (excluding 'LSOA code' and 'Names')
    correlations = df.drop(['LSOA code', 'Names'], axis=1).corrwith(df[n])

    # Display the correlations
    
    
    sorted_correlations = correlations.sort_values(ascending=False)

    # Create a bar plot of the sorted correlations
    plt.figure(figsize=(10, 10))
    sns.barplot(x=sorted_correlations.values, y=sorted_correlations.index, palette='coolwarm')
    plt.xlabel('Correlation')
    plt.ylabel('Columns')
    plt.title('Correlation between "count" and other columns for year ' + year)
    plt.show()

```

Corellation between the count of crimes in a year and different demographic staistics

```python
for year in range(2010, 2015) :
    corr_year(year)
```

```python
corr_year(2011)
```
